## NaN's: Checking

### Categorical

```python
def nulls_percentage(df):
    nulls=pd.DataFrame(df.isna().sum()*100/len(df), columns=['nulls_%'])
    nulls=nulls.sort_values('nulls_%', ascending = False)
    return nulls
```

```python
cat_nulls=nulls_percentage(cat_raw)
cat_nulls;
```

```python
num_nulls=nulls_percentage(num_raw)
num_nulls;
```

```python
all_nulls=nulls_percentage(all_raw)
all_nulls;
```

- The **only column with NaN's in the categorical data** is the column **GEOCODE2** with **~0.14% of nulls**.
- The majority of nulls are in the numerical data.

```python
drop_list=[]
drop_list.append('OSOURCE')
drop_list.append('ZIP')
drop_list   
```

```python
all_nulls_85=all_nulls[(all_nulls['nulls_%'] >= 85)]
all_nulls_85;
```

```python
(all_nulls_85.shape[0] / all_raw.shape[1])*100
```

- We've identified **87 columns** that have **more than 85% of nulls**. This represents **5.2% of the columns of the original dataset**.

```python
def remove_85(df):
    nulls=nulls_percentage(df)
    nulls_85=nulls[(nulls['nulls_%'] > 85)]
    for row in nulls_85.index:
        df = df.drop(row, axis=1)
    return df
```

```python
cat_raw_data=remove_85(cat_raw)
```

```python
cat_raw_data.shape
```

```python
cat_raw_data['GENDER'].unique()
```

```python
cat_raw_data['GENDER'] = cat_raw_data['GENDER'].fillna('F')
cat_raw_data['GENDER'] = cat_raw_data['GENDER'].replace(' ','OTHER')
cat_raw_data['GENDER'] = cat_raw_data['GENDER'].replace('C','OTHER')
cat_raw_data['GENDER'] = cat_raw_data['GENDER'].replace('U','OTHER')
cat_raw_data['GENDER'] = cat_raw_data['GENDER'].replace('J','OTHER')
cat_raw_data['GENDER'] = cat_raw_data['GENDER'].replace('A','OTHER')
```

```python
num_nulls=nulls_percentage(num_raw)
num_nulls
```

```python
def find_index(lst):
    dic={}
    for i in range(len(lst)):
        for j in range(len(num_nulls.index)):
            if num_nulls.index[j] == lst[i]:
                dic[lst[i]]=j
    return dic
```
